## Olasılığa Giriş ve Olasılık Dağılımları
* Rassal Değişken: Değerlerin, bir deneyin sonuçlarından alan değişken
* Dağılım: Olaylar veya durumların sayısal karşılıklarının ortaya çıkardığı yapı.
* Olasılık Dağılımı: Bir rassal olaya ait değerler ve bu değerlerin gerçekleşme olasılıklarının bir arada ifade edilmesi.
* Olasılık Fonksiyonu: Bir değişkenin herhangi bir değeri almasının olasılığını hesaplayan fonksiyon.

> ## Kesikli Olasılık Dağılımları
> * Bernoulli
> * Binom
> * Poisson

> ## Sürekli Olaslık Dağılımları
> * Normal Dağılım
> * Uniform Dağılım
> * Üstel Dağılım

### Bernoilli Dağılımı
İki sonuçlu olaylar ile ilgilenildiğinde kullanılan kesikli olasılık dağılımıdır.

${\displaystyle f(k;p)=p^{k}(1-p)^{1-k}\quad {\text{for }}k\in \{0,1\}}$

${\displaystyle \operatorname {E} \left(X\right)=up}$

$\operatorname {Var} {X}=pq=p(1-p)$

```python
from scipy.stats import bernoulli
p = 0.6
rv = bernoulli(p)
print(rv.pmf(k = 1), rv.pmf(k = 0))
```

### Büyük Sayılar Yasası
Bir rassal değişkenin uzun vadeli kararlılığını tanımlayan olasılık teoremi.

```python
rng = np.random.RandomState(123)
for i in np.arange(1, 21):
    deney_sayisi = 2**i
    yazi_turalar = rng.randint(0, 2, size = deney_sayisi)
    yazi_olasiliklari = np.mean(yazi_turalar)
    print("Atış SAyısı: ", deney_sayisi, "---", "Yazı olasılığı: %.2f" % (yazi_olasiliklari * 100))
```

## Binom Dağılımı
Bağımsız n deneme sonucu k başırılı olma olasılığı ile ilgilenildiğinde kullanılır.

```python
from scipy.stats import binom
p = 0.01
n = 100
rv = binom(n, p)
print("1 kişi: ", rv.pmf(1))
print("5 kişi: ", rv.pmf(5))
print("10 kişi: ", rv.pmf(10))
```

## Poisson Dağılımı
Belirli bir zaman aralığında belirli bir alanda nadir görülen olayların olasılıklarını hesaplama
* Rassal denemeler iki sonuçlu olmalı
* Aynı koşullar altında gerçekleşmeli
* Rassal denemeler birbirinden bağımsız olmalı

```python
from scipy.stats import poisson
lambda_ = 0.1
rv = poisson(mu = lambda_)
print("0 hata: ", rv.pmf(k = 0))
print("1 hata: ", rv.pmf(k = 1))
print("3 hata: ", rv.pmf(k = 3))
print("5 hata: ", rv.pmf(k = 5))
```

## Normal Dağılım
Normal dağıldığı bilinen, sürekli rassal değişkenler için kullanılır.

```python
from scipy.stats import norm
print("90'dan fazla: ", 1 - norm.cdf(90, 80, 5))
print("70'ten fazla: ", 1 - norm.cdf(70, 80, 5))
print("73'ten az: ", norm.cdf(73, 80, 5))
print("85 ile 90 arası: ", norm.cdf(90, 80, 5) - norm.cdf(85, 80, 5))
```
